```python
def isUpValid(row):
    return row > 0

def isDownValid(row):
    return row < 2

def isLeftValid(col):
    return col > 0

def isRightValid(col):
    return col < 2

def moveUp(state, space):
    row, col = space
    if isUpValid(row):
        new = [list(r) for r in state]
        new[row][col], new[row-1][col] = new[row-1][col], new[row][col]
        return tuple(tuple(r) for r in new)
    return states

def moveDown(state, space):
    row, col = space
    if isDownValid(row):
        new = [list(r) for r in state]
        new[row][col], new[row+1][col] = new[row+1][col], new[row][col]
        return tuple(tuple(r) for r in new)
    return state

def moveLeft(state, space):
    row, col = space
    if isLeftValid(col):
        new = [list(r) for r in state]
        new[row][col], new[row][col-1] = new[row][col-1], new[row][col]
        return tuple(tuple(r) for r in new)
    return state

def moveRight(state, space):
    row, col = space
    if isRightValid(col):
        new = [list(r) for r in state]
        new[row][col], new[row][col+1] = new[row][col+1], new[row][col]
        return tuple(tuple(r) for r in new)
    return state

def findSpace(state):
    for r in range(3):
        for c in range(3):
            if state[r][c] == 0:
                return (r, c)
    return None
```

```python
def getsuccessors(state):
    row, col = findSpace(state)
    successors = []
    
    if isUpValid(row):
        new_state = moveUp(state, (row, col))
        successors.append((new_state, "Up"))
    
    if isDownValid(row):
        new_state = moveDown(state, (row, col))
        successors.append((new_state, "Down"))
    
    if isLeftValid(col):
        new_state = moveLeft(state, (row, col))
        successors.append((new_state, "Left"))
    
    if isRightValid(col):
        new_state = moveRight(state, (row, col))
        successors.append((new_state, "Right"))
    
    return successors
```

```python
def print_state(state):
    for row in state:
        print(' '.join(map(str, row)))
    print()  
```

```python
def print_moves(states, moves):
    for state, move in zip(states, moves):
        print_state(state)
        print(f"Move: {move}")
    print_state(states[-1]) 
    print(f"Move: {moves[-1]}")  
```

```python
from collections import deque

def bfs(start, goal):
    if start == goal:
        print_state(start)
        return [start]
    
    queue = deque([(start, [], [])]) 
    visited = set()
    visited.add(start)
    
    while queue:
        state, path, moves = queue.popleft()
        
        for next, move in getsuccessors(state):
            if next in visited:
                continue
            
            newp = path + [state]
            newm = moves + [move]
            
            if next == goal:
                print_moves(newp + [next], newm + [move])
                return newp + [next]
            
            visited.add(next)
            queue.append((next, newp, newm))
    
    return None
```
